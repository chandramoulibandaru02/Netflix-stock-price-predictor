```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
data=pd.read_csv("NFLX.csv")
```

```python
data.head(10)
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

## Here there is no null values and hence we can do the prediction using linear regression

```python
## we have to split the data and take the target value 
```

## Hence we can't take date because the linear regression will only take the numeric values 

```python
x=data.drop(columns=['Date','Adj Close'])
```

```python
y=data['Adj Close']
```

## Now we have to split the data

```python
from sklearn.model_selection import train_test_split as t
```

```python
x_test,x_train,y_test,y_train=t(x,y,random_state=42,test_size=0.4)
```

## we have to check the shapes of the splitted data

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_test,y_test)
```

```python
y_pred=model.predict(x_train)
```

```python
print(y_pred)
```

## There fore they are the predicted values

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
acc=r2_score(y_train,y_pred)
```

```python
print(acc)
```

```python
abs=mean_absolute_error(y_train,y_pred)
```

```python
print(abs)
```

```python
me=mean_squared_error(y_train,y_pred)
```

```python
print(me)
```
